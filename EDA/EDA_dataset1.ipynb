```python
# Identificar outliers para algunas variables numéricas clave
num_columns = ['Customer_Age', 'Credit_Limit', 'Total_Trans_Amt', 'Avg_Utilization_Ratio']
outliers = {col: data[col].quantile([0.25, 0.75]) for col in num_columns}
iqr_outliers = {col: outliers[col].iloc[1] - outliers[col].iloc[0] for col in num_columns}

# Agregar límites superior e inferior para detectar outliers
outlier_limits = {
    col: (outliers[col].iloc[0] - 1.5 * iqr_outliers[col], outliers[col].iloc[1] + 1.5 * iqr_outliers[col])
    for col in num_columns
}

outlier_counts = {
    col: len(data[(data[col] < outlier_limits[col][0]) | (data[col] > outlier_limits[col][1])])
    for col in num_columns
}

outlier_counts
```

**Outliers:**
1. Customer_Age: 2 valores extremos detectados (fuera del rango esperado).
2. Credit_Limit: 984 registros tienen límites de crédito considerados outliers.
3. Total_Trans_Amt: 896 registros tienen cantidades totales de transacciones consideradas outliers.
4. Avg_Utilization_Ratio: No se detectaron outliers para esta variable.

```python
# Crear diagramas de caja para visualizar outliers
plt.figure(figsize=(16, 8))

for i, col in enumerate(num_columns, start=1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=data[col], palette="pastel")
    plt.title(f"Diagrama de Caja - {col}", fontsize=12)
    plt.xlabel(col, fontsize=10)
    plt.grid(axis="y")

plt.tight_layout()
plt.show()

# Calcular media y mediana con y sin outliers
stats_with_outliers = data[num_columns].describe().loc[['mean', '50%']]

# Filtrar datos excluyendo outliers
data_no_outliers = data.copy()
for col in num_columns:
    lower_limit, upper_limit = outlier_limits[col]
    data_no_outliers = data_no_outliers[(data_no_outliers[col] >= lower_limit) & (data_no_outliers[col] <= upper_limit)]

stats_without_outliers = data_no_outliers[num_columns].describe().loc[['mean', '50%']]

stats_with_outliers, stats_without_outliers
```

**Análisis del Impacto de los Outliers:**
1. Credit_Limit y Total_Trans_Amt: Muestran valores extremadamente altos que están muy por encima de los percentiles superiores.
2. Customer_Age: Presenta solo un par de valores fuera del rango.
3. Avg_Utilization_Ratio: No se identifican outliers significativos.

**Comparación de las medias y medianas:**

**Credit_Limit:**
1. Media con outliers: 8,631.95.
2. Media sin outliers: 5,918.24 (disminución significativa).
3. La mediana (4549 con outliers, 3624 sin outliers) es decir los outliers sesgan fuertemente la distribución hacia valores altos.

**Total_Trans_Amt:**
1. Media con outliers: 4,404.08.
2. Media sin outliers: 3,473.69 (reducción notable).
La mediana no cambia mucho, indicando que los outliers afectan principalmente la media.
**Customer_Age:**
1, La diferencia entre estadísticas con y sin outliers es mínima, lo que sugiere que el impacto es despreciable.
**Propuesta de Acción:**
En Credit_Limit y Total_Trans_Amt: los outliers tienen un impacto significativo.
**Acción para los outliers con alto impacto:** Aplicar transformación logarítmica o eliminación si los valores no son consistentes con el resto de los datos.
